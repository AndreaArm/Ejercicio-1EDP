La solución de la ecuación 
$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0 $$
sujeta a las condiciones de frontera:
$$u(0,y)=10y, ~~ u_x(1,y)=-1, \;\;\; 0 < y < 1$$
$$u(x,0)=u(x,1)=0\;\;\; 0 < x < 1 $$\
Con la Serie
$$ u(x,y) = \sum_{n=1}^{\infty} \sin(n \pi y) [ a_n \cosh (n \pi x) + b_n \sinh(n \pi x)] $$


```python
from pylab import *
from matplotlib import animation,rc
from IPython.display import HTML 
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
```

Utilizamos el valor obtenido en An=$$a_n=\frac{-20 (-1)^n}{n \pi}$$

Ynuestro valor obtenido en Bn=$$b_n=\frac{2[(-1)^n-1]+20(-1)^n(n \pi)\sinh (n\pi)}{(n \pi)^2 \cosh (n \pi)}$$

```python
def u_x_y(x,y):
    N = 8
    f = zeros ((len(x),len(y)))
    for d in range(len(x)):
        for n in range(1,N):
          a = (-20*(-1)**n)/(n*pi)
          b = (2*((-1)**n-1)+20*(n*pi)*sinh(n*pi)*(-1)**n)/(n**2*pi**2*cosh(n*pi))
          f[:,d] += sin(n*pi*y)*(a*cosh(n*pi*x[d])+b*sinh(n*pi*x[d]))
    return f
```

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="viridis"); #Otros colores:"viridis", "inferno", "plasma"
```
