La solución de la ecuación 
$$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2}$$
sujeta a las condiciones de frontera:
 0<x<1 o<y<t$$
Para:
$$ u(0,y)=u(1,y)=0; u(x,0)=100, u(x,1)=200

Esta dada en términos de la serie de Fourier por la expresión:


$$\sum_{n=1}^{\infty} [\frac{200(1-(-1))^n}{n\pi}(cosh(n\pi*y))+(\frac{200(1-(-1)^n(2-cosh(n\pi)}{senh(n\pi)+(n\pi)^2})senh(yn\pi)]sen(xn\pi) .$$

```python
from pylab import *
from matplotlib import animation,rc
from IPython.display import HTML 
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
x= linspace(0,1,1000)
y= linspace(0,1,1000)
```

Utilizaremos nuestra serie de fourier separando el valor de An y Bn

```python
def u_x_y(x,y):
    N = 10
    f = zeros ((len(x),len(y)))
    for i in range(len(x)):
        for n in range(1,N):
          a = (-20*(-1)^n)/(n*pi)
          b = (2*((-1)**n-1)+20*(-1)**n*(n*pi)*sinh(n*pi))/((n*pi)**2*cosh(n*pi))
          f[:,i] += sin(n*pi*y)*(a*cosh(n*pi*x[i])+b*sinh(n*pi*x[i]))
    return f
```

```python

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 

ax.plot_surface(Y, X, U, cmap="viridis"); #Otros colores:"viridis", "inferno", "spectral"
```
